```python
def rating (value):
    if value <= 2:
        return 'низкий рейтинг'
    elif value <= 4:
        return 'средний рейтинг'
    elif 4.5 <= value <= 5:
        return 'высокий рейтинг'
    return 'неправильные данные рейтинга'
```

```python
import pandas as pd

file_ratings = pd.read_csv('/Users/macbook_bd/Downloads/ml-latest-small/ratings.csv')
file_movies = pd.read_csv('/Users/macbook_bd/Downloads/ml-latest-small/movies.csv')
common_file = pd.merge(file_movies, file_ratings, on='movieId', how='outer')
common_file['class'] = common_file['rating'].apply(rating)
common_file.groupby('class').count().reset_index()
```

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def region (value):
    value.lower()
    for sep_value in value.split():
        if sep_value in geo_data['Центр']:
            return 'Центр' 
        if sep_value in geo_data['Северо-Запад']:
            return 'Северо-Запад' 
        if sep_value in geo_data['Дальний Восток']:
            return 'Дальний Восток' 
    return 'undenied'
```

```python
import pandas as pd

file_keywords = pd.read_csv('/Users/macbook_bd/Downloads/ml-latest-small/keywords.csv')
file_keywords['regions'] = file_keywords['keyword'].apply(region)
```

```python
file_keywords.groupby('regions').count().reset_index()
```
