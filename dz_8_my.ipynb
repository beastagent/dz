```python
import json
import requests

class ExchangeRate:
    def __init__(self, format_ = 'Value', diff_ = 'False'):
        self.format = format_
        self.diff = diff_
    def return_results_json(self):
        self.take_results = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.take_results.json()['Valute']
    def choice_of_format(self, currency):
        valutes = self.return_results_json()
        if currency in valutes:
            if self.diff == 'False':
                if self.format == 'Value':
                    return valutes[currency] ['Value']
                elif self.format == 'Full':
                    return valutes[currency]
                return "Вводи для формата ответа 'Full' или 'Value'"
            elif self.diff == 'True':
                return round((valutes[currency]['Value'] - valutes[currency]['Previous'])/valutes[currency]['Previous'] * 100, 2)
            return "Вводи для параметра разницы 'True' или 'False'"
    def CNY(self):
        return self.choice_of_format('CNY')
    def EUR(self):
        return self.choice_of_format('EUR')
    def USD(self):
        return self.choice_of_format('USD')
    def BYN(self):
        return self.choice_of_format('BYN')
                
```

```python
a1 = ExchangeRate(format_ = 'Full', diff_ = 'False')
a1.BYN()
```

```python
a2 = ExchangeRate(format_ = 'Value')
a2.BYN()
```

```python
a3 = ExchangeRate(format_ = 'Full', diff_ = 'True')
a3.BYN()
```
