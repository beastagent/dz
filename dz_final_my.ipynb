```python
def production_year (row):
    years = list(range(1950,2011))
    if re.search(r'(\d{4})',row['title']):
        if int(re.search(r'(\d{4})',row['title']).group()) in  years:
            return int(re.search('(\d{4})',row['title']).group())
    return 1900
```

```python
import pandas as pd
import re
table = pd.read_csv('Downloads/movies_stats.csv')

table['years'] = table.apply(production_year, axis=1)
new_table = table.groupby('years').agg({'rating': 'mean'}).reset_index().sort_values(by='rating', ascending = False)
new_table.head(100)
```
