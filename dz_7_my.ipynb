```python
newspapers_dates = ('The Moscow Times — Wednesday, October 2, 2002',
'The Guardian — Friday, 11.10.13',
'Daily News — Thursday, 18 August 1977')

from datetime import datetime

print(datetime.strptime(newspapers_dates[0], 'The Moscow Times — %A, %B %d, %Y'))
print(datetime.strptime(newspapers_dates[1], 'The Guardian — %A, %d.%m.%y'))
print(datetime.strptime(newspapers_dates[2], 'Daily News — %A, %d %B %Y'))
```

```python
from datetime import datetime
from datetime import timedelta

def list_of_dates(start_date:str, end_date:str):
    try:
        dates_list = []
        s_d = datetime.strptime(start_date, '%Y-%m-%d')
        e_d = datetime.strptime(end_date, '%Y-%m-%d')
        while s_d <= e_d:
            dates_list.append(s_d.strftime('%Y-%m-%d'))
            s_d += timedelta(days=1)
        if dates_list == []:
            return('Неправильно введены данные. Даты вводятся в порядке возрастания')
        return dates_list
    except :
        return('Неправильно введены данные. Даты должны вводиться в формате \'yyyy-mm-dd\'')
        
try:
    print(list_of_dates('2025-3-03', '2025-03-25'))
except:
    print('Неправильно введены данные. Даты должны вводиться в формате \'yyyy-mm-dd\'')
```
